```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import numpy as np
```

```python
book = pd.read_csv("C:/Users/srina/Desktop/csv files/book.csv")
book.head()
```

#### Creating Frequent Itemset with min_support=0.01

```python
frequent_itemsets1 = apriori(book, min_support=0.01, use_colnames=True)
frequent_itemsets1['length'] = frequent_itemsets1['itemsets'].apply(lambda x : len(x))
frequent_itemsets1
```

#### Creating Frequent Itemset with min_support=0.02

```python
frequent_itemsets2 = apriori(book, min_support=0.02, use_colnames=True)
frequent_itemsets2['length'] = frequent_itemsets2['itemsets'].apply(lambda x : len(x))
frequent_itemsets2
```

#### Creating Frequent Itemset with min_support=0.03

```python
frequent_itemsets3 = apriori(book, min_support=0.03, use_colnames=True)
frequent_itemsets3['length'] = frequent_itemsets3['itemsets'].apply(lambda x : len(x))
frequent_itemsets3
```

```python
frequent_itemsets1[frequent_itemsets1['length'] >= 2].head(20)
```

```python
frequent_itemsets2[frequent_itemsets2['length'] >= 2].head(20)
```

```python
frequent_itemsets3[frequent_itemsets3['length'] >= 2].head(20)
```

#### Creating Rule for the Frequent Itemset with min_support=0.01

```python
rules1 = association_rules(frequent_itemsets1, metric="lift", min_threshold=0.7)
rules1
```

#### Creating Rules for the Frequent Itemset with min_support=0.02

```python
rules2 = association_rules(frequent_itemsets2, metric="lift", min_threshold=0.7)
rules2
```

#### Creating Rule for the Frequent Itemset with min_support=0.03

```python
rules3 = association_rules(frequent_itemsets3, metric="lift", min_threshold=0.7)
rules3
```

#### An leverage value of 0 indicates independence. Range will be [-1 1]
A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

### Scatter Plot for Rule1

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10,6)
color = plt.cm.inferno(np.linspace(0,1,20))
rules1.value_counts().head(20).plot.bar(color = color)
plt.title('Top 20 Most Frequent Items')
plt.ylabel('Counts')
plt.xlabel('Items')
plt.show()
```
